## Noise Removal

```python
# The code will remove the noisy words from a text
```

```python
noise_list = ['is','the','a','this','with','that','in','as']
```

```python
from nltk.corpus import stopwords
stop_words = list(stopwords.words('english'))
stop_words
```

```python
def remove_noise(input_text):
    words = input_text.split()
    noise_free_words = [w for w in words if w not in stop_words]
    new_text = ' '.join(noise_free_words)
    return new_text
    
```

## Stemming and Lemmatization

```python
# Play, Playing, Played ----> play
```

```python
from nltk.stem import PorterStemmer, SnowballStemmer
```

```python
ps = PorterStemmer()
ss = SnowballStemmer(language='english')
```

```python
words= ['Play','Playing','Played','Playable','Happened','Happening','Lovingly','sportingly']
```

```python
for w in words:
    print(w,"  ",ps.stem(w,to_lowercase=True),"  ",ss.stem(w.lower()))
```

```python
def stemming(input_text):
    words = word_tokenize(input_text)
    noise_free_words = [w for w in words if w not in stop_words]
    stem_words = [ss.stem(w.lower()) for w in noise_free_words]
    final_text = ' '.join(stem_words)
    return final_text
```

```python
stemming('I like playing when its raining')
```

```python
stemming('My experience with Techstack was tremendous. The Faculty is very friendly & cooperative. They give theoretical and also practical knowledge of each subject. The course fee was also affordable and I enjoyed every moment spent learning here. Forever grateful for who I am today!')
```

```python
# WAP to remove the hashtags from twitter comment using regex
```

```python
def rem_hashtags (input_text,pattern):
    hashtags = re.findall(pattern,input_text)
    for i in hashtags:
                  input_text = re.sub(i,"",input_text)
    return input_text
```

```python
rem_hashtags('Its high time that you all Wake Up! #HindusUnderAttackInIndia #DelhiRiots #HinduLivesMatters','#[\w]*')
```

## Object Standardization

```python
dict = {'ty':'thank you',
       'idk':'I dont know',
       'ttyl': 'talk to you later',
       'luv': 'love',
       'thx':'thanks'}
```

```python
def lookup_words(input_text):
    words = word_tokenize(input_text)
    new_text = []
    for w in words:
        if w.lower() in dict:
            w = dict[w.lower()]
        new_text.append(w)
    new_sent = " ".join(new_text)
    return new_sent
```

```python
lookup_words('ty so much, idk how to say thx, lot of luv to u, ttyl')
```

# N Grams

```python
def gen_ngrams(text, n):
    terms = word_tokenize(text)
    print(terms)
    new_terms = [spl_terms for spl_terms in terms if spl_terms not in stop_words]
    output = []
    for i in range(len(new_terms)-n+1):
        output.append(new_terms[i:i+n])
    return output
```

## Cosine Similarity

```python
import math
from collections import Counter
from nltk.tokenize import word_tokenize
```

```python
def text_vec(text):
    word = word_tokenize(text)
    return Counter(word)
```

```python
text_vec('hello I am ravi kumar')
```

```python
def cos_sim(text1, text2):
    vec1 = text_vec(text1)
    vec2 = text_vec(text2)
    
    com = set(vec1.keys())& set(vec2.keys())
    nume= sum(vec1[i]*vec2[i] for i in com)
    print(com,nume)
    
    deno1 = sum(vec1[i]**2 for i in vec1.keys())
    deno2 = sum(vec2[i]**2 for i in vec2.keys())
    
    deno = math.sqrt(deno1)*math.sqrt(deno2)
    print(deno)
    
    cosine = nume/deno
    print('Cosine Similarity: ', cosine)
```

```python
cos_sim('Hi I am ravi','Hello I am suman')
```

```python
cos_sim('It is raining in delhi','I like to eat cheese')
```
